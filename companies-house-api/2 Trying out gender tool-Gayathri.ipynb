```python
def get_save_unique_first_names(officers_df, 
                                first_name_file='all_officers_first_names'):
    unique_first_names = officers_df[0].str.split(
        ', ', expand=True)[1].unique()
    pd.DataFrame(unique_first_names).to_csv(first_name_file, index=False)
```

```python
def get_officers_df_with_first_name(officers_df):
    with_first_name = pd.concat(
        [officers_df, officers_df[0].str.split(
               ', ', expand=True)[1]],
          axis=1)
    with_first_name = with_first_name.rename(
        index=str, columns={1: "officer_first_name"})
    return with_first_name
```
