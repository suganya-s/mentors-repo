```python
def get_save_unique_first_names(officers_df, 
                                first_name_file='all_officers_first_names'):
    unique_first_names = officers_df['Names_Active_Officers'].str.split(
        ', ', expand=True)[1].unique()
    pd.DataFrame(unique_first_names).to_csv(first_name_file, index=False)
    
def get_save_unique_first_names_2(officers_df, 
                                first_name_file='active_officers_first_names'):
    unique_first_names = officers_df['first_name'].unique()
    pd.DataFrame(unique_first_names).to_csv(first_name_file, index=False)
```
